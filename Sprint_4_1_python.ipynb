```python
import sqlite3
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
def make_query(query, filepath='database.sqlite'):
    """
    Execute a query on a SQLite database file and return the results as a pandas DataFrame.

    Parameters
    ----------
    query : str
        SQL query to be executed on the database file.
    filepath : str, optional
        Path to the database file, by default 'database.sqlite'.

    Returns
    -------
    pd.DataFrame
        A pandas DataFrame containing the results of the executed query.
    """
    with sqlite3.connect(filepath) as connection:
        cursor = connection.cursor()
        execution = cursor.execute(query)
        result = pd.DataFrame(execution.fetchall(), columns=[i[0] for i in execution.description])
        return result
```

```python
player_attributes = make_query(
"""
SELECT *
FROM player_attributes
""")

player = make_query(
"""
SELECT *      
FROM player
""")

match = make_query(
"""
SELECT *      
FROM match
""")

team = make_query(
"""
SELECT *      
FROM team
""")

team_attributes = make_query(
"""
SELECT *      
FROM team_attributes
""")

league = make_query(
"""
SELECT *      
FROM league
""")

country = make_query(
"""
SELECT *      
FROM country
""")

display(player, player_attributes, league, match, team, team_attributes, country)
display(player.columns, player_attributes.columns, league.columns, match.columns, 
        team.columns, team_attributes.columns, country.columns)
```

```python
def Match(x):
    columns = ['home_player_' + str(i) for i in range(1,12)] + ['away_player_' + str(i) for i in range(1, 12)]
    pl = player_attributes.groupby(['player_api_id']).mean().round(0).reset_index()
    dictio = dict(zip(pl['player_api_id'], pl[x]))
    
    for kol in columns:
        column_name = kol + '_rating'
        match[column_name] = [dictio.get(a, None) for a in match[kol]]
    return match
Match('overall_rating').filter(like="player").mean
```

```python
#Opdracht 4.3
import matplotlib.pyplot as plt
import seaborn as sns
# Filter matches for 2015-2016 season
matches = match[match['season'] == '2008/2009']

# Merge matches with team to get team names
matches_teams = matches.merge(team[['team_api_id', 'team_long_name']], left_on='home_team_api_id', right_on='team_api_id', how='inner')
matches_teams2 = matches_teams.rename(columns={'team_long_name': 'thuis_team_name'})
matches_teams3 = pd.merge(matches_teams2, team[['team_api_id', 'team_long_name']], left_on='away_team_api_id', right_on='team_api_id', how='inner')
matches_teams4 = matches_teams3.rename(columns={'team_long_name': 'uit_team_name'})

matches_teams4
player_match = matches_teams4.merge(player_attributes, on='date', how='inner')

playername_match = player_match.merge(player, on = 'player_api_id', how = 'inner' )
playername_match
playername_match_short = playername_match[["date", "season", "match_api_id", "player_name", "player_fifa_api_id_y"
                                           , "overall_rating", "team_api_id_x", "thuis_team_name", "team_api_id_y"
                                           ,"uit_team_name"]]
team_ratings = playername_match_short.groupby(['thuis_team_name'])['overall_rating'].mean().reset_index()
team_ratings = team_ratings.append(playername_match_short.groupby(['uit_team_name'])['overall_rating'].mean().reset_index())
team_ratings = team_ratings.groupby(['thuis_team_name'])['overall_rating'].mean().reset_index()

# Create bar plot of team ratings
plt.figure(figsize=(15, 8))
sns.barplot(x="thuis_team_name", y="overall_rating", data=team_ratings)
plt.title("Average rating per team voor the 2008-2009")
plt.xlabel("Team")
plt.ylabel("gemiddelde rating")
plt.xticks(rotation=90)
plt.show()
```
